```python
iris = pd.read_table('Iris.csv', sep=",",header = 0,index_col=0)
di = pd.DataFrame(data = iris)
print(di.describe())
```

```python
sns.pairplot(di)
```

```python
from sklearn . cluster import KMeans
import random
random.seed(10)
# Cluster K - means
model = KMeans(n_clusters =2,random_state = 0)
# adapter le m o d l e de d o n n e s
model.fit(iris[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']])
res_label = model.labels_
res_label
```

```python
# On ajoute une colonne qui contiendra le cluster choisi par kmeans
iris.insert(5,"kmeansClass", model.labels_,True)
iris.head()
```

```python
import seaborn as sns
sns.lmplot(x='SepalLengthCm', y='SepalWidthCm',data=iris,  hue ='kmeansClass', fit_reg = False)
plt.show()
```
